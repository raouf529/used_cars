```python
import pandas as pd
df = pd.read_csv("pre-owned cars.csv")
brand_models = df.groupby("brand")["model"].unique().to_dict()
```

```python
brand_models
```

```python
import json
with open("brand_models.json", "w") as f:
    json.dump({k: list(v) for k, v in brand_models.items()}, f, indent=2)
```
